```python
import plotly.express as px
df = df_train
# Here we use a column with categorical data
px.histogram(df, x="HomePlanet", color='Transported')
# fig.show()
```

```python
df=df_train
cs_t_grp = df.groupby(['CryoSleep','HomePlanet','Transported'])['PassengerId'].count()
cs_t_grp = cs_t_grp.reset_index().rename(columns={'PassengerId':'Number of Passengers'})
display(cs_t_grp)
px.bar(data_frame=cs_t_grp,x='HomePlanet',y='Number of Passengers',color='Transported',pattern_shape='CryoSleep',labels={
    'x':'CryoSleep',
},title='Number of Passengers Tranported vs CryoSleep')
```
